```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math
```

```python
pco2 = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (0), skip_header = 1)
temp = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (1), skip_header = 1)
ec = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (2), skip_header = 1)
pges = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (3), skip_header = 1)
o2 = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (4), skip_header = 1)
toc = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (5), skip_header = 1)
cat = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (6), skip_header = 1)
an = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (7), skip_header = 1)
pH = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (8), skip_header = 1)
date = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (9), skip_header = 1)
```

#### pCO2 vs Time

```python
dateint = np.zeros(len(date))
for i in range(len(date)):
    dateint[i] = int(date[i])



jan = 0
feb = 0
mar = 0
apr = 0
mai = 0
jun = 0
jul = 0 
aug = 0
sep = 0
octo = 0
nov = 0
dec = 0

for i in range(len(date)):
    if dateint[i] == 1:
        jan = jan + 1
    if dateint[i] == 2:
        feb = feb + 1
    if dateint[i] == 3:
        mar = mar + 1
    if dateint[i] == 4:
        apr = apr + 1
    if dateint[i] == 5:
        mai = mai + 1
    if dateint[i] == 6:
        jun = jun + 1
    if dateint[i] == 7:
        jul = jul + 1
    if dateint[i] == 8:
        aug = aug + 1
    if dateint[i] == 9:
        sep = sep + 1
    if dateint[i] == 10:
        octo = octo + 1
    if dateint[i] == 11:
        nov = nov + 1
    if dateint[i] == 12:
        dec = dec + 1


january = np.zeros(jan)
february = np.zeros(feb)
march = np.zeros(mar)
april = np.zeros(apr)
may = np.zeros(mai)
june = np.zeros(jun)
july = np.zeros(jul)
august = np.zeros(aug)
september = np.zeros(sep)
october = np.zeros(octo)
november = np.zeros(nov)
december = np.zeros(dec)


def monthsort(month,monthno,monthct=0):
    ''' sorts the data in each month into a separate array '''
    for i in range(len(dateint)):
        if dateint[i] == monthno:
            month[monthct] = pco2[i]
            monthct = monthct + 1
    return month

january = monthsort(january,1)
february = monthsort(february,2)
march = monthsort(march,3)
april = monthsort(april,4)
may = monthsort(may,5)
june = monthsort(june,6)
july = monthsort(july,7)
august = monthsort(august,8)
september = monthsort(september,9)
october = monthsort(october,10)
november = monthsort(november,11)
december = monthsort(december,12)


data_to_plot = [january, february, march, april, may, june, july, august, september, october, november, december]

fig = plt.figure(18)
ax = fig.add_subplot(111)
bp = ax.boxplot(data_to_plot)
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr','May', 'Jun', 'Jul', 'Aug','Sep', 'Oct', 'Nov', 'Dec'])
ax.get_xaxis().tick_bottom()
ax.set_title('pCO2 vs time (months)')
plt.xlabel('Months')
plt.ylabel('pCO2 [ppm]')
```

```python
def getkey(item):
    return item[0]
```

#### Water Temperature

```python
datatemp = np.zeros((len(pco2),2))
for i in range(len(pco2)):
    datatemp[i,0] = temp[i]
    datatemp[i,1] = pco2[i]

datatemp2 = sorted(datatemp,key=getkey)

tempordpco2 = np.zeros(len(pco2))
ordtemp = np.zeros(len(pco2))
for i in range(len(pco2)):
    tempordpco2[i] = datatemp2[i][1]
    ordtemp[i] = datatemp2[i][0]

tempbinpco2 = np.zeros(int(len(pco2)/5.0))
bintemp = np.zeros(len(tempbinpco2))
tempstdpco2 = np.zeros(len(tempbinpco2))
for j in range(len(tempbinpco2)):    
    for i in range(len(pco2)):
        if i == ((j+1)*5):
            tempbinpco2[j] = np.mean(np.array((tempordpco2[i-5],tempordpco2[i-4],tempordpco2[i-3],tempordpco2[i-2],tempordpco2[i-1])))
            bintemp[j] = np.median(np.array((ordtemp[i-5],ordtemp[i-4],ordtemp[i-3],ordtemp[i-2],ordtemp[i-1])))
            tempstdpco2[j] = np.std(np.array((tempordpco2[i-5],tempordpco2[i-4],tempordpco2[i-3],tempordpco2[i-2],tempordpco2[i-1])))
        if i == (len(pco2)-1.0) and j == len(tempbinpco2)-1:
            tempbinpco2[j] = np.mean(np.array((tempordpco2[i-3],tempordpco2[i-2],tempordpco2[i-1], tempordpco2[i])))
            bintemp[j] = np.median(np.array((ordtemp[i-3],ordtemp[i-2],ordtemp[i-1],ordtemp[i])))
            tempstdpco2[j] = np.std(np.array((tempordpco2[i-3],tempordpco2[i-2],tempordpco2[i-1], tempordpco2[i])))


slope, intercept, r_value, p_value, std_err = stats.linregress(bintemp,tempbinpco2)
print 'Water temp R^2 =', r_value**2
print 'Water temp p value =',p_value

plt.figure(0)
plt.plot(temp,pco2,'o')
plt.xlabel('Water Temp [deg C]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs Water Temperature, Unbinned')

plt.figure(1)
plt.errorbar(bintemp,tempbinpco2,tempstdpco2,linestyle='None',marker='o')
plt.plot(np.linspace(bintemp[0],bintemp[len(bintemp)-1]),slope*np.linspace(bintemp[0],bintemp[len(bintemp)-1])+intercept)
plt.xlabel('Water Temp [deg C]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs Water Temperature, Binned')
```

```python
datapH = np.zeros((len(pco2),2))
for i in range(len(pco2)):
    datapH[i,0] = pH[i]
    datapH[i,1] = pco2[i]

datapH2 = sorted(datapH,key=getkey)

pHordpco2 = np.zeros(len(pco2))
ordpH = np.zeros(len(pco2))
for i in range(len(pco2)):
    pHordpco2[i] = datapH2[i][1]
    ordpH[i] = datapH2[i][0]

pHbinpco2 = np.zeros(int(len(pco2)/5.0))
binpH = np.zeros(len(pHbinpco2))
pHstdpco2 = np.zeros(len(pHbinpco2))
for j in range(len(pHbinpco2)):    
    for i in range(len(pco2)):
        if i == ((j+1)*5):
            pHbinpco2[j] = np.mean(np.array((pHordpco2[i-5],pHordpco2[i-4],pHordpco2[i-3],pHordpco2[i-2],pHordpco2[i-1])))
            binpH[j] = np.median(np.array((ordpH[i-5],ordpH[i-4],ordpH[i-3],ordpH[i-2],ordpH[i-1])))
            pHstdpco2[j] = np.std(np.array((pHordpco2[i-5],pHordpco2[i-4],pHordpco2[i-3],pHordpco2[i-2],pHordpco2[i-1])))
        if i == (len(pco2)-1.0) and j == len(pHbinpco2)-1:
            pHbinpco2[j] = np.mean(np.array((pHordpco2[i-3],pHordpco2[i-2],pHordpco2[i-1], pHordpco2[i])))
            binpH[j] = np.median(np.array((ordpH[i-3],ordpH[i-2],ordpH[i-1],ordpH[i])))
            pHstdpco2[j] = np.std(np.array((pHordpco2[i-3],pHordpco2[i-2],pHordpco2[i-1], pHordpco2[i])))


logpHbinpco2 = np.zeros(len(pHbinpco2))
for i in range(len(pHbinpco2)):
    logpHbinpco2[i] = math.log(pHbinpco2[i])

pHslope, pHintercept, pHr_value, pHp_value, pHstd_err = stats.linregress(binpH,logpHbinpco2)


logpHfit = np.zeros(len(pHbinpco2))
for i in range(len(pHbinpco2)):
    logpHfit[i] = pHslope*binpH[i]+pHintercept

explogpHbinpco2 = np.zeros(len(pHbinpco2))
for i in range(len(pHbinpco2)):
    explogpHbinpco2[i] = math.exp(logpHfit[i])



print 'pH logscale R^2 =', pHr_value**2
print 'pH logscale p value =',pHp_value


plt.figure(16)
plt.plot(pH,pco2,'o')
plt.yscale('log')
plt.xlabel('pH')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs pH, Unbinned') 

plt.figure(17)
plt.errorbar(binpH,pHbinpco2,pHstdpco2,linestyle='None',marker='o')
plt.plot(binpH,explogpHbinpco2)
plt.yscale('log')
plt.xlabel('pH')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs pH, Binned') 
```
