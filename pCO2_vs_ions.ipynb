In this notebook, we plot pCO2 vs the other ions: Na+, K+, Ca+2, Mg+2, NO2-, Cl-, and SO4-2. We also plot them specific to location.

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
pco2 = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (0), skip_header = 1)
na = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (11), skip_header = 1)
k = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (12), skip_header = 1)
ca = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (13), skip_header = 1)
mg = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (14), skip_header = 1)
no2 = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (15), skip_header = 1)
cl = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (16), skip_header = 1)
so4 = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (17), skip_header = 1)

temp = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (1), skip_header = 1)
ec = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (2), skip_header = 1)
pges = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (3), skip_header = 1)
o2 = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (4), skip_header = 1)
toc = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (5), skip_header = 1)
cat = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (6), skip_header = 1)
an = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (7), skip_header = 1)
pH = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (8), skip_header = 1)
alk = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (10), skip_header = 1)
```

```python
def removenan(ion):
    '''finds and removes the nan values from the data'''
    
    mask = ~np.isnan(ion) & ~np.isnan(pco2)
    length = len(ion[mask])    
    
    data = np.zeros((2,length))
    for i in range(length):    
        data[0,i] = ion[mask][i]
        data[1,i] = pco2[mask][i]
        
    return data
```

```python
def locsep(index1,index2):
    length = index2-index1+1
    data = np.zeros((17,length))
    for i in range(length):
        data[0,i]  = pco2[i+index1]
        data[1,i]  = temp[i+index1]
        data[2,i]  = ec[i+index1]
        data[3,i]  = pges[i+index1]
        data[4,i]  = o2[i+index1]
        data[5,i]  = toc[i+index1]
        data[6,i]  = cat[i+index1]
        data[7,i]  = an[i+index1]
        data[8,i]  = pH[i+index1]
        data[9,i]  = alk[i+index1]
        data[10,i]  = na[i+index1]
        data[11,i] = k[i+index1]
        data[12,i] = ca[i+index1]
        data[13,i] = mg[i+index1]
        data[14,i] = no2[i+index1]
        data[15,i] = cl[i+index1]
        data[16,i] = so4[i+index1]
    return data


nahe1 = locsep(0,11)
nahe2 = locsep(12,24)
nahe3 = locsep(25,37)
nahe4 = locsep(38,50)
nahe5 = locsep(51,63)
nahe6 = locsep(64,85)
nahe7 = locsep(86,98)
nahe8 = locsep(99,111)
nahe9 = locsep(112,124)
nahe10 = locsep(125,137)
nahe11 = locsep(138,150)
nahe12 = locsep(151,163)
nahe13 = locsep(164,176)



ngw1 = locsep(177,188)
ngw2 = locsep(189,200)
ngw3 = locsep(201,212)
ngw4 = locsep(213,224)
ngw5 = locsep(225,236)
ngw6 = locsep(237,248)
ngw7 = locsep(249,260)
ngw8 = locsep(261,272)
ngw9 = locsep(273,284)
ngw10 = locsep(285,296)
ngw11 = locsep(297,308)
ngw12 = locsep(309,320)
ngw13 = locsep(321,332)
ngw14 = locsep(333,344)
ngw15 = locsep(345,356)
ngw16 = locsep(357,368)



rhein1 = locsep(369,380)
rhein2 = locsep(381,392)
rhein3 = locsep(393,404)
rhein4 = locsep(405,416)
rhein5 = locsep(417,428)
rhein6 = locsep(429,440)
rhein7 = locsep(441,452)
rhein8 = locsep(453,464)
rhein9 = locsep(465,476)
rhein10 = locsep(477,488)




def plotpco2(varnum,xlabel,varname,subnum):
    plt.subplot(7,2,subnum)
    plt.plot(nahe1[varnum],nahe1[0],'o',color='Cyan',ms=4)
    plt.plot(nahe2[varnum],nahe2[0],'o',color='PaleTurquoise',ms=4)
    plt.plot(nahe3[varnum],nahe3[0],'o',color='Turquoise',ms=4)
    plt.plot(nahe4[varnum],nahe4[0],'o',color='PowderBlue',ms=4)
    plt.plot(nahe5[varnum],nahe5[0],'o',color='SteelBlue',ms=4)
    plt.plot(nahe6[varnum],nahe6[0],'o',color='DeepSkyBlue',ms=4)
    plt.plot(nahe7[varnum],nahe7[0],'o',color='SkyBlue',ms=4)
    plt.plot(nahe8[varnum],nahe8[0],'o',color='DodgerBlue',ms=4)
    plt.plot(nahe9[varnum],nahe9[0],'o',color='RoyalBlue',ms=4)
    plt.plot(nahe10[varnum],nahe10[0],'o',color='MediumBlue',ms=4)
    plt.plot(nahe11[varnum],nahe11[0],'o',color='MidnightBlue',ms=4)
    plt.plot(nahe12[varnum],nahe12[0],'o',color='LightSlateGray',ms=4)
    plt.plot(nahe13[varnum],nahe13[0],'o',color='MediumSlateBlue',ms=4)
    plt.plot(ngw1[varnum],ngw1[0],'o',color='IndianRed',ms=4)
    plt.plot(ngw2[varnum],ngw2[0],'o',color='LightCoral',ms=4)
    plt.plot(ngw3[varnum],ngw3[0],'o',color='Salmon',ms=4)
    plt.plot(ngw4[varnum],ngw4[0],'o',color='DarkSalmon',ms=4)
    plt.plot(ngw5[varnum],ngw5[0],'o',color='LightSalmon',ms=4)
    plt.plot(ngw6[varnum],ngw6[0],'o',color='Crimson',ms=4)
    plt.plot(ngw7[varnum],ngw7[0],'o',color='Red',ms=4)
    plt.plot(ngw8[varnum],ngw8[0],'o',color='Firebrick',ms=4)
    plt.plot(ngw9[varnum],ngw9[0],'o',color='DarkRed',ms=4)
    plt.plot(ngw10[varnum],ngw10[0],'o',color='HotPink',ms=4)
    plt.plot(ngw11[varnum],ngw11[0],'o',color='DeepPink',ms=4)
    plt.plot(ngw12[varnum],ngw12[0],'o',color='Coral',ms=4)
    plt.plot(ngw13[varnum],ngw13[0],'o',color='Tomato',ms=4)
    plt.plot(ngw14[varnum],ngw14[0],'o',color='OrangeRed',ms=4)
    plt.plot(ngw15[varnum],ngw15[0],'o',color='Maroon',ms=4)
    plt.plot(ngw16[varnum],ngw16[0],'o',color='PaleVioletRed',ms=4)
    plt.plot(rhein1[varnum],rhein1[0],'o',color='GreenYellow',ms=4)
    plt.plot(rhein2[varnum],rhein2[0],'o',color='Lime',ms=4)
    plt.plot(rhein3[varnum],rhein3[0],'o',color='LimeGreen',ms=4)
    plt.plot(rhein4[varnum],rhein4[0],'o',color='PaleGreen',ms=4)
    plt.plot(rhein5[varnum],rhein5[0],'o',color='Green',ms=4)
    plt.plot(rhein6[varnum],rhein6[0],'o',color='DarkGreen',ms=4)
    plt.plot(rhein7[varnum],rhein7[0],'o',color='YellowGreen',ms=4)
    plt.plot(rhein8[varnum],rhein8[0],'o',color='OliveDrab',ms=4)
    plt.plot(rhein9[varnum],rhein9[0],'o',color='Olive',ms=4)
    plt.plot(rhein10[varnum],rhein10[0],'o',color='DarkOliveGreen',ms=4)
    plt.xlabel(xlabel)
    plt.ylabel('pCO2 [ppm]')
    plt.title('pCO2 vs '+varname)
```

```python
plt.figure(figsize=(16,35))


plotpco2(10,'Na [mg/l]','Na',2)
plotpco2(11,'K [mg/l]','K',4)
plotpco2(12,'Ca [mg/l]','Ca',6)
plotpco2(13,'Mg [mg/l]','Mg',8)
plotpco2(14,'NO2 [mg/l]','NO2',10)
plotpco2(15,'Cl [mg/l]','Cl',12)
plotpco2(16,'SO4 [mg/l]','SO4',14)



plt.subplot(7,2,1)
plt.plot(na,pco2,'o')
plt.xlabel('Na [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs Na')




slopek, interceptk, r_valuek, p_valuek, std_errk = stats.linregress(removenan(k)[0],removenan(k)[1])
print "K R^2 =", r_valuek**2
print "K p value =", p_valuek

klin = np.linspace(min(k),max(k))
plt.subplot(7,2,3)
plt.plot(k,pco2,'o',klin,slopek*klin+interceptk)
plt.xlabel('K [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs K')





slopeca, interceptca, r_valueca, p_valueca, std_errca = stats.linregress(removenan(ca)[0],removenan(ca)[1])
print "Ca R^2 =", r_valueca**2
print "Ca p value =", p_valueca

calin = np.linspace(min(ca),max(ca))

plt.subplot(7,2,5)
plt.plot(ca,pco2,'o',calin,slopeca*calin+interceptca)
plt.xlabel('Ca [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs Ca')





slopemg, interceptmg, r_valuemg, p_valuemg, std_errmg = stats.linregress(removenan(mg)[0],removenan(mg)[1])
print "Mg R^2 =", r_valuemg**2
print "Mg p value =", p_valuemg

mglin = np.linspace(min(mg),max(mg))

plt.subplot(7,2,7)
plt.plot(mg,pco2,'o',mglin,slopemg*mglin+interceptmg)
plt.xlabel('Mg [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs Mg')




slopeno2, interceptno2, r_valueno2, p_valueno2, std_errno2 = stats.linregress(no2,pco2)
print "NO2 R^2 =", r_valueno2**2
print "NO2 p value =", p_valueno2

no2lin = np.linspace(min(no2),max(no2))

plt.subplot(7,2,9)
plt.plot(no2,pco2,'o',no2lin,slopeno2*no2lin+interceptno2)
plt.xlabel('NO2 [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs NO2')




plt.subplot(7,2,11)
plt.plot(cl,pco2,'o')
plt.xlabel('Cl [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs Cl')




plt.subplot(7,2,13)
plt.plot(so4,pco2,'o')
plt.xlabel('SO4 [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs SO4')
```
