In this notebook, we plot pCO2 vs time, water temperature, electrical conductivity, phosophorus, oxygen, TOC, total cations, total anions, and pH. We also plot TOC vs phosphorus. Each of these sets of plotted data are plotted with the original data as well as binned versions and with outlying points removed.

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
pco2 = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (0), skip_header = 1)
na = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (11), skip_header = 1)
k = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (12), skip_header = 1)
ca = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (13), skip_header = 1)
mg = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (14), skip_header = 1)
no2 = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (15), skip_header = 1)
cl = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (16), skip_header = 1)
so4 = np.genfromtxt("pCO2_etc.txt", delimiter = '\t', usecols = (17), skip_header = 1)
```

```python
def removenan(ion):
    '''finds and removes the nan values from the data'''
    
    mask = ~np.isnan(ion) & ~np.isnan(pco2)
    length = len(ion[mask])    
    
    data = np.zeros((2,length))
    for i in range(length):    
        data[0,i] = ion[mask][i]
        data[1,i] = pco2[mask][i]
        
    return data
```

```python
plt.figure()
plt.plot(na,pco2,'o')
plt.xlabel('Na [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs Na')




slopek, interceptk, r_valuek, p_valuek, std_errk = stats.linregress(removenan(k)[0],removenan(k)[1])
print "K R^2 =", r_valuek**2
print "K p value =", p_valuek

klin = np.linspace(min(k),max(k))
plt.figure()
plt.plot(k,pco2,'o',klin,slopek*klin+interceptk)
plt.xlabel('K [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs K')





slopeca, interceptca, r_valueca, p_valueca, std_errca = stats.linregress(removenan(ca)[0],removenan(ca)[1])
print "Ca R^2 =", r_valueca**2
print "Ca p value =", p_valueca

calin = np.linspace(min(ca),max(ca))

plt.figure()
plt.plot(ca,pco2,'o',calin,slopeca*calin+interceptca)
plt.xlabel('Ca [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs Ca')





slopemg, interceptmg, r_valuemg, p_valuemg, std_errmg = stats.linregress(removenan(mg)[0],removenan(mg)[1])
print "Mg R^2 =", r_valuemg**2
print "Mg p value =", p_valuemg

mglin = np.linspace(min(mg),max(mg))

plt.figure()
plt.plot(mg,pco2,'o',mglin,slopemg*mglin+interceptmg)
plt.xlabel('Mg [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs Mg')




slopeno2, interceptno2, r_valueno2, p_valueno2, std_errno2 = stats.linregress(no2,pco2)
print "NO2 R^2 =", r_valueno2**2
print "NO2 p value =", p_valueno2

no2lin = np.linspace(min(no2),max(no2))

plt.figure()
plt.plot(no2,pco2,'o',no2lin,slopeno2*no2lin+interceptno2)
plt.xlabel('NO2 [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs NO2')




plt.figure()
plt.plot(cl,pco2,'o')
plt.xlabel('Cl [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs Cl')




plt.figure()
plt.plot(so4,pco2,'o')
plt.xlabel('SO4 [mg/l]')
plt.ylabel('pCO2 [ppm]')
plt.title('pCO2 vs SO4')

plt.show()
```
